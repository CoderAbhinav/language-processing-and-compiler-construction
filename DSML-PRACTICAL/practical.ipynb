```python
def ic(nums: list) -> float:
    import math
    s, res = sum(nums), 0

    for i in nums:
        res += (i / s) * math.log2(i / s)
    
    return (- 1) * res
```

```python
def info(data: pd.DataFrame,param: str) -> float:
    import math
    n, res = data[param].size, 0
    nums = data[param].value_counts().to_list()    
    return ic(nums)
```

```python
def infoA(data: pd.DataFrame,d: str, a: str) -> float:
    unique_vals_in_a = data[a].value_counts().to_dict()
    unique_vals_in_d = data[d].value_counts().to_dict()

    n , res = data[d].size, 0

    for i, j in unique_vals_in_a.items():
        x = ic(data[data[a] == i][d].value_counts().to_list())
        res += (j / n) * x

    return res
```

```python
def gain(data: pd.DataFrame,d: str, a: str) -> float:
    return info(data, d) - infoA(data, d, a)
```

```python
def dataSetInfoGain(data: pd.DataFrame, a: str):
    for i in data.columns.to_list():
        gi = gain(data, a, i)
        print(f"Gain({i}) = {gi}")
```

```python
class DescisionTreeNode:
    def __init__(self, attb: str, isLeaf: bool  = True, descision: str = "") -> None:
        self.__descision = descision
        self.__isLeaf = isLeaf
        self.__attb: str = attb
        self.__label: dict = {}


    @property
    def attribute_value(self) -> str:
        return self.__attb
    
    @property
    def labels(self) -> dict:
        return self.__label
    
    def addLabel(self, label, val):
        self.__label[label] = val

    @property
    def isLeaf(self) -> bool:
        return self.__isLeaf
    
    @property
    def descision(self) -> str:
        return self.__descision
    

class DescisionTreeBuilder:
    def ic(self, nums: list) -> float:
        import math
        s, res = sum(nums), 0

        for i in nums:
            res += (i / s) * math.log2(i / s)
        
        return (- 1) * res
    
    def info(self, data: pd.DataFrame,param: str) -> float:
        import math
        n, res = data[param].size, 0
        nums = data[param].value_counts().to_list()    
        return self.ic(nums)

    def infoA(self, data: pd.DataFrame,d: str, a: str) -> float:
        unique_vals_in_a = data[a].value_counts().to_dict()
        unique_vals_in_d = data[d].value_counts().to_dict()

        n , res = data[d].size, 0

        for i, j in unique_vals_in_a.items():
            x = self.ic(data[data[a] == i][d].value_counts().to_list())
            res += (j / n) * x

        return res
    
    def gain(self, data: pd.DataFrame,d: str, a: str) -> float:
        return self.info(data, d) - self.infoA(data, d, a)
    
    def dataSetInfoGain(self, data: pd.DataFrame, a: str) -> dict:
        res = {}
        for i in data.columns.to_list():
            if i == a:
                continue
            
            gi = self.gain(data, a, i)

            res[i] = gi

        return res

    
    def nodeBuilder(self, data: pd.DataFrame, a: str) -> DescisionTreeNode:
        if (data[a].value_counts().size == 1):
            return DescisionTreeNode(a, True, data[a].unique()[0])

        gains: dict = self.dataSetInfoGain(data, a)
        max_info_gain = max(gains, key=gains.get)

        atbs = data[max_info_gain].unique()

        curr: DescisionTreeNode = DescisionTreeNode(max_info_gain)

        print(max_info_gain ,atbs)
        
        for i in atbs:
            temp = data[data[max_info_gain] == i]
            temp.drop(columns=max_info_gain)
            curr.addLabel(i, self.nodeBuilder(temp, a))
            
        

        return curr

    def __init__(self, data: pd.DataFrame, target: str) -> None:
        self.__data : pd.DataFrame = data
        self.__target : str = target
        self.__root : DescisionTreeNode = self.nodeBuilder(data, target)
    
    @property
    def getData(self) -> pd.DataFrame:
        return self.__data

    @property
    def getRootNode(self) -> DescisionTreeNode:
        return self.__root
    

    
```

Perform the k-mean clustering with initial centroids as

Answer the following

```python
import pandas as pd
from matplotlib import pyplot as plt


data = pd.DataFrame({'point':['P1','P2','P3','P4','P5','P6','P7','P8'],'x' : [0.1,0.15,0.08,0.16,0.2,0.25,0.24,0.3],'y' : [0.6,0.71,0.9,0.85,0.3,0.5,0.1,0.2]})
```

```python
pts = list(zip(data["x"].values, data["y"].values))
```

```python
m1 = (0.1, 0.6)
m2 = (0.3, 0.2)
c1 = []
c1d = []
c2 = []
c2d = []

for i in pts:
    d1 = eucledianDistance(m1, i)
    d2 = eucledianDistance(m2, i)

    if d1 < d2:
        c1.append(i)
        c1d.append(d1)
    else:
        c2.append(i)
        c2d.append(d2)

c1x = [i for i, j in c1]
c1y = [j for i, j in c1]

c2x = [i for i, j in c2]
c2y = [j for i, j in c2]
```

```python
def clustering(data: list, m1: int, m2: int) -> tuple:
    m1 = data[m1]
    m2 = data[m2]

    c1 = []
    c2 = []

    c1d = []
    c2d = []

    def eucledianDistance(p1: tuple, p2: tuple) -> float:
        x1, y1 = p1
        x2, y2 = p2

        return abs(x1 - x2) + abs(y1 - y2)

    for i in data:
        d1 = eucledianDistance(m1, i)
        d2 = eucledianDistance(m2, i)

        if d1 > d2:
            c1.append(i)
            c1d.append(d1)
        else:
            c2.append(i)
            c2d.append(d2)


    return (c1, c2, c1d, c2d)
    
    
```

```python
for i in range(0, 5):
    print(f"Centroids are P{i + 1} = {pts[i]}, P{len(pts) - i} = {pts[len(pts) - i - 1]}")
    c1, c2, c1d, c2d = clustering(pts, i, len(pts) - i - 1)

    print(c1, c2)
    print(sum(c1d))
    print(sum(c2d))
    print(sum(c1d) + sum(c2d))
```
